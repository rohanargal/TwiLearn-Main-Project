```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Airbnb data.csv')
```

```python
NULL = df.isna().sum()
NULL
```

```python
## We see,that there are 4 empty columns in dataset which is relatable to filling the missing values taking reference of other columns
```

```python
df.columns
```

```python
df['country'].fillna(value='Netherland', axis= 'index', inplace=True)
```

```python
df['borough'].fillna(value= 'Centrum', axis= 'index', inplace=True)
```

```python
df['minstay'].fillna(value= '1-Day',axis='index', inplace=True)
```

```python
df['bathrooms'].fillna(value= '1', axis='index', inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.rename(columns= {'country':'Country', 'borough':'Borough', 'bathrooms':'Bathrooms', 'minstay':'Minstay', 'room_id':'Room_Id', 'survey_id':'Survey_Id', 'host_id':'Host_Id', 'room_type':'Room_type', 'city':'City', 'reviews':'Reviews', 'overall_satisfaction':'Overall_satisfaction', 'accommodates':'Accommodates', 'bedrooms':'Bedrooms', 'price':'Price', 'name':'Name', 'last_modified':'Last_modified', 'latitude':'Latitude', 'longitude':'Longitude', 'location':'Location'}, inplace=True)
```

```python
df.dtypes
```

```python
# 1.Which is the highest earning.
high=df.groupby(['Name', 'Host_Id'])['Price'].mean().reset_index().sort_values(by='Price', ascending=False)
high
```

```python
# Answer 1:- 
#Top three earners.
#Zonnige woonboot,centraal en rustig : 6000.0
#One public bedroom : 3770.0
#AmsterdamBase : 1920.0
```

```python
# 2. Is Any particular location getting a maximum number of bookings.
maxi =df.groupby(['Location', 'neighborhood'])['Reviews'].sum().reset_index().sort_values(by='Reviews', ascending=False)
maxi
```

```python
plt.figure(figsize=(22,8))
sns.scatterplot(data=maxi, x='neighborhood', y='Reviews', palette= 'hsv')
plt.xticks(rotation=45)
plt.title("Scatterplot")
plt.xlabel('Neighborhood')
plt.tight_layout()
plt.ylabel('Reviews')
plt.show()
```

```python
# Answer 2:- 
# Top Three Maximum Bookings.
#Centrum West: 532
#De Baarsjes / Oud West: 465
#Centrum West: 463
```

```python
# 3.Price relation with respect to location.
PR_LA = df.groupby(['neighborhood'])['Price'].mean().reset_index().sort_values(by='Price', ascending=False)
PR_LA
```

```python
plt.figure(figsize=(22,10))
sns.barplot(data=PR_LA, x='neighborhood', y='Price', palette='magma')
plt.xticks(rotation=45)
plt.title("Swarmplot")
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.show()
```

```python
#Answer 3:- As we check to center of city we observe that their is an increase in Price because of population and availability & it also cameout that Centrum west has highest price & 2nd highest avg price.
```

```python
df.head()
```

```python
# 4. Relationship between Quality and Price
Quality = df.groupby(['Overall_satisfaction', 'neighborhood'])['Price'].mean().reset_index().sort_values(by='Overall_satisfaction', ascending=False)
Quality
```

```python
plt.figure(figsize=(22,8))
sns.boxplot(data=Quality, x='neighborhood', y='Price', palette='viridis')
plt.title("Scatterplot")
plt.xticks(rotation=90)
plt.ylabel('Overall_satisfaction')
plt.xlabel('Price')
plt.show()
```

```python
# Answer 4:-
# We can see that increase in quality by refering overall_satisfaction of hotels which are in center of city with avg price rate.
```

```python
#Price vs amenities
Top_20 = df.groupby(['Overall_satisfaction','Name'])['Price'].mean().reset_index().sort_values(by='Price', ascending=False)[:20]
Top_20
```

```python
plt.figure(figsize=(22,8))
ax = plt.axes()
ax.set_facecolor('Black')
sns.scatterplot(data=Top_20, x='Name', y='Price', color='red', s=200, edgecolor='black', marker= '^')
plt.title("Barplot")
plt.xticks(rotation=45)
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.show()
```

```python
#Price vs location

PR_LOC = df.groupby(['Location', 'neighborhood'])['Price'].mean().reset_index().sort_values(by='Price', ascending=False)
PR_LOC
```

```python
plt.figure(figsize=(22,8))
sns.barplot(data=PR_LOC, x='neighborhood', y='Price', palette='viridis')
plt.title("Barplot")
plt.xticks(rotation=45)
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.show()
```
