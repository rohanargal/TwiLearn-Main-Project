```python
# Converting "Order Date & Ship Date" to Datetime Type.
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

```python
# Bar plot by Country and total revenue.
plt.figure(figsize=(12, 6))
plt.bar(df['Country'], df['Total Revenue'])
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Units Sold', y='Total Revenue', data=df)
plt.xlabel('Units Sold')
plt.ylabel('Total Revenue')
plt.title('Units Sold vs. Total Revenue')
plt.show()
```

```python
# Visualize Units Sold by Item Type
plt.figure(figsize=(8, 6))
sns.boxplot(x='Item Type', y='Units Sold', data=df)
plt.xticks(rotation=90)
plt.title('Units Sold by Item Type')
plt.show()
```

```python
# As per required objective Data-sets to be Needed in Months, Year, Yearly-Month.
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Year_Month'] = df['Order Date'].dt.to_period('M')
```

```python
# Group the data by month and calculate total revenue for each month
sales_monthly = df.groupby('Month')['Total Revenue'].sum()
sales_monthly
```

```python
# Plotting the sales trend month-wise
plt.figure(figsize=(12, 6))
sales_monthly.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Sales Trend - Month-wise')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#grouping data by Year & calculating Total revenue by Yearly.
sales_yearly = df.groupby('Year')['Total Revenue'].sum()
sales_yearly
```

```python
# Plotting the sales trend year-wise
plt.figure(figsize=(12, 6))
sales_yearly.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.title('Sales Trend - Year-wise')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group the data by year-month and calculate total revenue for each year-month
sales_yearly_monthly = df.groupby('Year_Month')['Total Revenue'].sum()
sales_yearly_monthly
```

```python
# Plotting the sales trend yearly-monthly
plt.figure(figsize=(12, 6))
sales_yearly_monthly.plot(kind='line')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.title('Sales Trend - Yearly-Month-wise')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
